```python
import matplotlib.pyplot as plt
import csv

x = []
y1 = []
y2 = []

with open('patentTimeDependetData.tsv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        x.append(int(row[0]))
        y1.append(int(row[1]))
        y2.append(int(row[2]))
        
plt.plot(x,y1, label='Patent (Same)')
plt.plot(x,y2, label='Patent (Different)')

plt.xlabel('Year')
plt.ylabel('# of patents')

plt.legend(loc="lower center", bbox_to_anchor=(0.5, -0.6))
plt.show()
```
