```python
import matplotlib.pyplot as plt
import csv

x = []
y = []

with open('graph2.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))

plt.plot(x,y, label='Loaded from file!')
plt.xlabel('Publication year')
plt.ylabel('Average number of genes with Same-Classification')
plt.title('Average number of genes with Same-Classification per publication')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

x = []
y = []
y2 = []

with open('graph3.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))
        y2.append(int(row[2]))

plt.plot(x,y, label='Publications with Same-Classification')
plt.plot(x,y2, label='Publications with Different-Classification')
plt.xlabel('Year')
plt.ylabel('# of publications')
plt.title('# of publications by year')
plt.legend()
plt.show()
# add overal number of pubMed publication
```

```python
import matplotlib.pyplot as plt
import csv

x = []
y = []
y2 = []
y3 = []

with open('graph4.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))
        y2.append(int(row[2]))
        y3.append(int(row[3]))

plt.plot(x,y, label='Publications with Same-Classification')
plt.plot(x,y2, label='Publications with Different-Classification')
plt.plot(x,y3, label='PubMed Publications')
plt.xlabel('Year')
plt.ylabel('# of publications')
plt.title('# of publications by year')
plt.legend()
plt.show()
# add overal number of pubMed publication
```
