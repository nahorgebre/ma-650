```python
still_multis = []

for m in mts:
    
    mtcopy = list(mts)
    
    mtcopy.remove(m)
    
    if m[0] in [x[0] for x in mtcopy]:
        still_multis.append(m)
        print(m[-1])
```

```python

oldest_notes = []
for s in still_multis:
    
    others = list(still_multis)
    others.remove(s)
    
    same_as_s = [x for x in others if x[0] == s[0]]
    print(same_as_s)
    
    first_note = s
    
    for y in same_as_s:
        if y[-1] < first_note[-1]:
            # y is older than first_note
            first_note = y
            
    oldest_notes.append(first_note)
```

```python
oldest_notes = list(set(oldest_notes))
```
