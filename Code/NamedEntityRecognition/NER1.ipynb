```python
import pandas as pd
import csv

abstracts_df = pd.read_csv('.\data\input\Abstract_test_100K.csv')
abstracts_df.head()
```

```python
ex_text = abstracts_df["abstract_text"][1]
ex_text
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag

pr_text = word_tokenize(ex_text)
pr_text = nltk.pos_tag(pr_text)

pr_text
```

```python
pattern = 'NP: {<DT>?<JJ>*<NN>}'
cp = nltk.RegexpParser(pattern)
cs = cp.parse(pr_text)
print(cs)
```

```python
from nltk.chunk import conlltags2tree, tree2conlltags
from pprint import pprint

iob_tagged = tree2conlltags(cs)
pprint(iob_tagged)
```
